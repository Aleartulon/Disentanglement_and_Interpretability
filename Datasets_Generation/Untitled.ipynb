```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sample_weird_shape(X):
    x = X[0]
    y = X[1]
    first_center = (0.3,0.3)
    second_center = (0.5,0.3)
    radius = 0.4
    length = 0.6
    if (x-first_center[0])**2+(y-first_center[1])**2 < 0.1**2 or (x-second_center[0])**2+(y-second_center[1])**2 < 0.1**2 or (x<second_center[0] and x>first_center[0] and (y-first_center[1])<length and (y>first_center[1])):
        return 1.0
    else:
        return 0.0
```

```python
space_samples = np.random.rand(10000,2)
print(np.shape(space_samples))
Y = []
for i in space_samples:
    y = sample_weird_shape(i)
    Y.append(y)
```

```python
plt.scatter(space_samples[:,0], space_samples[:,1], c = Y)
```
